#### **Importing Necessary Libraries**

```python
import pandas as pd
import matplotlib.pyplot as plt
import os
from itertools import combinations
from collections import Counter
```

**Merging 12 months of sales data into a single CSV file**

```python
files = []

for file in os.listdir('./Sales_Data'):
    if file.endswith('.csv'):
        files.append('./Sales_Data/' + file)

all_months_df = pd.concat(map(pd.read_csv, files), ignore_index = True)

all_months_df.to_csv("all_months.csv", index = False)
```

**Reading updated csv file into a dataframe**

```python
all_data_df = pd.read_csv('all_months.csv')
all_data_df.head()
```

```python
all_data_df.shape
```

#### **Cleaning up and prepping the data**

```python
nan_data_df = all_data_df[all_data_df.isna().any(axis=1)]
nan_data_df.head()
```

```python
all_data_df = all_data_df.dropna(how='all')
all_data_df.head()
```

**Removing rows whose Order Date column starts with 'Or'**

```python
all_data_df = all_data_df[all_data_df["Order Date"].str[0:2] != ("Or")]
all_data_df.head()
```

**Converting Quantity Ordered & Price Each to numeric datatypes**

```python
all_data_df['Quantity Ordered'] = pd.to_numeric(all_data_df['Quantity Ordered'])
all_data_df['Price Each'] = pd.to_numeric(all_data_df['Price Each'])
all_data_df.head()
```

**Adding a month column**

```python
all_data_df["Month"] = all_data_df["Order Date"].str[0:2]

all_data_df["Month"] = all_data_df["Month"].astype('int32')
all_data_df.head()
```

**Adding a Sales column**

```python
all_data_df['Sales'] = all_data_df['Quantity Ordered'] * all_data_df['Price Each']
all_data_df = all_data_df[["Order ID","Product","Quantity Ordered", "Price Each", "Sales", "Order Date", "Purchase Address", "Month"]]
all_data_df.head()
```

**Adding a City column**

```python
def get_city(address):
    return address.split(',')[1]

def get_state(address):
    return address.split(',')[2].split(' ')[1]

all_data_df['City'] = all_data_df['Purchase Address'].apply(lambda x: f"{get_city(x)} ({get_state(x)})")
all_data_df.head()
```

#### ****What was the best month for sales? How much was earned that month?****

```python
monthly_sales = all_data_df.groupby('Month').sum()['Sales']
monthly_sales
```

```python
months = range(1,13)

plt.bar(months, monthly_sales)
plt.xticks(months)
plt.ylabel('Sales in USD')
plt.xlabel('Month number')
plt.show()
```

#### ****What city had the highest sales?****

```python
sales_by_city = all_data_df.groupby('City').sum()['Sales']
sales_by_city
```

```python
cities = [city for city, df in all_data_df.groupby('City')] #list comprehension

plt.bar(cities, sales_by_city)
plt.xticks(cities, rotation='vertical', size=10)
plt.ylabel('Sales in USD')
plt.xlabel('City')
plt.show()
```

#### **What is the best time to display advertisements to maximize the likelihood of customers buying the product?**

```python
all_data_df.head()
```

```python
all_data_df['Order Date'] = pd.to_datetime(all_data_df['Order Date'], format='mixed')
all_data_df.head()
```

```python
all_data_df['Hour'] = all_data_df['Order Date'].dt.hour
all_data_df['Minute'] = all_data_df['Order Date'].dt.minute
all_data_df.head()
```

```python
hours = [hour for hour, df in all_data_df.groupby('Hour')]

plt.plot(hours, all_data_df.groupby('Hour').count())
plt.xticks(hours)
plt.xlabel('Hours')
plt.ylabel('No. of Orders')
plt.grid()
plt.show()

#We see peaks around 12pm and 7pm. So we need to run ads right before that.
```

#### **What products are often sold together?**

```python
dup_df = all_data_df[all_data_df['Order ID'].duplicated(keep=False)]
dup_df.head()
```

```python
dup_df['Grouped'] = dup_df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
dup_df = dup_df[['Order ID', 'Grouped']]
dup_df.head()
```

```python
dup_df = dup_df.drop_duplicates(subset=['Order ID'])
dup_df.head()
```

```python
count = Counter()

for row in dup_df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))

count.most_common(10)
```

#### **What product is sold the most? And why?**

```python
all_data_df.head()
```

```python
product_quantity = all_data_df.groupby('Product')['Quantity Ordered'].sum()
products = [product for product, df in all_data_df.groupby('Product')]

plt.bar(products, product_quantity)
plt.xticks(products, rotation='vertical')
plt.xlabel('Product Name')
plt.ylabel('Quantity Ordered')
plt.show()

#AAA Batteries is the product that was sold the most.
```

```python
prices = all_data_df.groupby('Product')['Price Each'].mean()

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()

ax1.bar(products, product_quantity, color='b')
ax2.plot(products, prices, 'r-')

ax1.set_xlabel('Product Name')
ax1.set_ylabel('Quantity Ordered', color='b')

ax2.set_xlabel('Product Name')
ax2.set_ylabel('Price in USD', color='r')

ax1.set_xticks(products)
ax1.set_xticklabels(products, rotation='vertical')

plt.show()
#We can see a negative correlation between the prices and quantity ordered.
```
